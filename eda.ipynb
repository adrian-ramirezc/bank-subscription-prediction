### Segment the target by numeric features

```python
num_cols = ['age', 'duration', 'campaign', 'pdays','previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

fig, ax =plt.subplots(3,3, figsize=(20,20))

k = 0
for i in range(3):
    for j in range(3):        
        sns.boxplot(y = df['y'].astype('category'), x = num_cols[k], data=df, ax=ax[i][j])
        k = k + 1
```

### Interactions between  numeric features and other numeric features

```python
import numpy as np
conditions  = [(df['y'] == 'yes'),(df['y'] == 'no')]
values      = [1,0]

df["y_num"] = np.select(conditions,values) 
df.sample(15)
```
